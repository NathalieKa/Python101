## Aufgabe: Grundlegende Statistik in Python

In dieser Aufgabe verwenden wir die Python-Bibliotheken **NumPy** und **Matplotlib**, um eine Stichprobe zu erzeugen, zu analysieren und und zu visualisieren.

Wir verwenden ein realitätsnahes Beispiel – die **Körpergröße von Personen** – und berechnen wichtige statistische Kennzahlen wie **Mittelwert**, **Median** und die **Standardabweichung**. Anschließend stellen wir die Verteilung visuell dar und interpretieren Ihre Ergebnisse.

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Aufgabe: Stichprobe erzeugen**

Erzeugen Sie eine Zufallsstichprobe mit 500 Werten aus einer **Normalverteilung** mit dem Mittelwert \( \mu = 170 \) und der Standardabweichung \( \sigma = 8 \). Verwenden Sie dafür die Funktion `np.random.normal`.

Weisen Sie die generierten Werte der Variable `data` zu.

```python
import numpy as np

data = np.random.normal(loc=170, scale = 8, size = 500)
#raise NotImplementedError()
```

```python
# Test Zelle
assert 'data' in globals(), "Variable 'data' wurde nicht definiert"
assert isinstance(data, np.ndarray), "'data' muss ein NumPy-Array sein"
assert len(data) == 500, "Die Stichprobe muss genau 500 Werte enthalten"
```

**Aufgabe: Statistische Kennzahlen**

Berechnen Sie auf Basis der generierten Stichprobe `data` folgende Kennzahlen:

- Arithmetisches Mittel (mean)
- Median
- Varianz
- Standardabweichung
- Interquartilsabstand (IQR)

Speichern Sie die Ergebnisse in den Variablen `mean_val`, `median_val`, `var_val`, `std_val` und `iqr_val`.

```python
mean_val = np.mean(data)   # Variable für den Mittelwert
median_val = np.median(data)  # Variable für den Median
var_val = np.var(data)     # Variable für die Varianz
std_val = np.std(data)     # Variable für die Standardabweichungd

# YOUR CODE HERE
#raise NotImplementedError()
```

```python
# Test Zelle
assert all(name in globals() for name in [
    "mean_val", "median_val", "var_val", "std_val"
]), "Mindestens eine der geforderten Variablen fehlt."
assert isinstance(mean_val, (float, np.float64)), "Mittelwert sollte ein float sein."
```

**Aufgabe: Daten visualisieren**

Erstellen Sie zwei Darstellungen der Stichprobe:

1. Ein Histogramm zur Darstellung der Verteilung
2. Einen Boxplot zur Visualisierung von Lage und Streuung

Nutzen Sie `matplotlib.pyplot` und stellen Sie die beiden Plots nebeneinander in einem gemeinsamen Plot-Fenster dar.

```python

import matplotlib.pyplot as plt

#fig = Figurm , ax1/ax2 = Zahlenbereiche
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5)) #Die Funktion plt.subplots() erzeugt ein Plot-Fenster (Figure) mit einem oder mehreren Zeichenbereichen (Axes).

# Verwende 20 bins
num_bins = 20

# Plotte Histogramm über data
ax1.hist(data, num_bins)

# Die x-Achse soll mit "Wert" gelabelt sein
ax1.set_xlabel("Wert")

# Die y-Achse soll das Label "Häufigkeit" tragen
ax1.set_ylabel("Häufigkeit")

# Der Title des Plots ist "Histogramm der Stichprobe"
ax1.set_title("Histogramm der Stichprobe")

# Plotte Boxplot über data
ax2.boxplot(data, vert=False)
# Der Plot soll horizontal (nicht vertikal) sein

# Die x-Achse soll das Label "Wert" tragen
ax2.set_xlabel("Wert")

# Der Title des Plots ist "Boxplot der Stichprobe"
ax2.set_title("Boxplot der Stichprobe")


# YOUR CODE HERE
#raise NotImplementedError()
```
